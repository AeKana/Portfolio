#### Identifying the distribution of Pokemon Types

Looking at number of Type 1 Pokemon:

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
type1_count=sns.countplot('Type 1', data=pokemon_df, palette=pkmn_palette1)
type1_count.set_xticklabels(type1_count.get_xticklabels(), rotation=40, ha="right")
plt.title('Number of Pokemon by Type 1 with Mega Evolution')
plt.show()
```

Now looking at at Type 2 Pokemon:

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
type2_count=sns.countplot('Type 2', data=pokemon_df, palette=pkmn_palette2)
type2_count.set_xticklabels(type2_count.get_xticklabels(), rotation=40, ha="right")
plt.title('Number of Pokemon by Type 2 with Mega Evolution')
plt.show()
```

Great! Next we'll take a look at each Generation through 1-6 and see the different type distribution:

#### Pokemon Type by base stats

We'll identify whether or not certain Pokemon types are stronger by their base stats:

* HP
* Attack
* Defense
* Sp. Atk
* Sp. Def
* Speed

Legendary Pokemon are known to be the strongest and have an overall stat higher than the average Pokemon. Let's see the distribution based on type and if there are  some Pokemon rivaling their total stats.

```python
plt.figure(figsize=(12,6))
total_type=sns.swarmplot(x='Type 1',y='Total',data=pokemon_df,hue='Legendary',palette='coolwarm')
total_type.set_xticklabels(total_type.get_xticklabels(), rotation=40, ha="right")
plt.axhline(pokemon_df['Total'].mean(),color='red')
plt.title('Total Stats by Type 1')
plt.show()
```

```python
plt.figure(figsize=(12,6))
total_type2=sns.swarmplot(x='Type 2',y='Total',data=pokemon_df,hue='Legendary',palette='coolwarm')
total_type2.set_xticklabels(total_type2.get_xticklabels(), rotation=40, ha="right")
plt.axhline(pokemon_df['Total'].mean(),color='red')
plt.title('Total Stats by Type 2')
plt.show()
```

We'll identify if different generations of Pokemon produced stronger stats:

```python
# Dropping unnecessary indexes and finding the means per Generation

pokemon_df=pokemon_df.drop(['Total','#','Legendary'],axis=1)
gen_stats=pokemon_df.groupby('Generation').mean()
```

```python
plt.plot(gen_stats,marker='o')

fig=plt.gcf()
fig.set_size_inches(12,6)
plt.legend(gen_stats)
plt.title('Base Stats comparison per Generation')
plt.xlabel('Generation')
plt.ylabel('Base Value')

plt.show()
```

For a more accurate description with numbers: 

```python
plt.figure(figsize=(10,6))
sns.heatmap(gen_stats,annot=True,fmt='g')
plt.show()
```

Let's look at a quick correlation for our Generation stats:

```python
plt.figure(figsize=(10,6))
sns.heatmap(gen_stats.corr(),annot=True)
plt.show()
```

Finally, we'll look at if certain types have advantages in base stats:

```python
# Dropping unnecessary indexes and finding the means per Type

pokemon_df=pokemon_df.drop(['Generation'],axis=1)
type_stats=pokemon_df.groupby('Type 1').mean()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(type_stats,annot=True,fmt='g')
plt.show()
```

Correlation for our Type 1:

```python
plt.figure(figsize=(10,6))
sns.heatmap(type_stats.corr(),annot=True)
plt.show()
```

Correlation for our Type 2:

```python
plt.figure(figsize=(10,6))
sns.heatmap(type_stats2.corr(),annot=True)
plt.show()
```
